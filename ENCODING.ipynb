## OneHotEncoder

- In this technique, the categorical parameters will prepare separate columns for both Male and Female labels. So, wherever there is Male, the value will be 1 in Male column and 0 in Female column, and vice-versa. Letâ€™s understand with an example: Consider the data where fruits and their corresponding categorical values and prices are given.

```python
## importing OneHot Encoding
from sklearn.preprocessing import OneHotEncoder
```

```python
## creating an Object for OneHotEncoder
OE = OneHotEncoder(categories='auto',drop='first',sparse=False)
```

```python
OE.fit(X_train)
```

```python
X_tr_encoded = OE.transform(X_train)
```

```python
X_ts_encoded = OE.transform(X_test)
```

# OrdinalEncoder

- In ordinal encoding, each unique category value is assigned an integer value.   
- This is called an ordinal encoding or an integer encoding and is easily reversible. Often, integer values starting at zero are used.

```python
## importing ordinal encoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
## creating an object for OrdinalEncoder
OE  = OrdinalEncoder()
```

```python
## fitting the train data
OE.fit(X_train)
```

```python
#transforming the train data
X_OE = OE.transform(X_train)
```

```python
## traansforming thee test data
X_t_OE = OE.transform(X_test)
```

```python
X_OE
```

```python
X_t_OE
```

## LabelEcoder

- Label Encoder is used for encoding 1d data only                
- Label Encoding refers to converting the labels into a numeric form so as to convert them into the machine-readable form. Machine learning algorithms can then decide in a better way how those labels must be operated. It is an important pre-processing step for the structured dataset in supervised learning.

```python
## importing Label Encoder
from sklearn.preprocessing import LabelEncoder
```

## BinaryEncoding

- Binary encoding is a combination of Hash encoding and one-hot encoding. In this encoding scheme, the categorical feature is first converted into numerical using an ordinal encoder. Then the numbers are transformed in the binary number. After that binary value is split into different columns.

```python
from category_encoders import BinaryEncoder
```

```python
BE = BinaryEncoder()
```

```python
X_BE = BE.fit_transform(X_train)
```

```python
X_t_BE = BE.transform(X_test)
```

```python
X_BE
```

```python
X_t_BE
```

# Base N Encoding

- In the numeral system, the Base or the radix is the number of digits or a combination of digits and letters used to represent the numbers. The most common base we use in our life is 10  or decimal system as here we use 10 unique digits i.e 0 to 9 to represent all the numbers. Another widely used system is binary i.e. the base is 2. It uses 0 and 1 i.e 2 digits to express all the numbers.

```python
from category_encoders import BaseNEncoder
```

```python
BN = BaseNEncoder(return_df=True,base=7)
```

```python
X_BN = BN.fit_transform(X_train)
```

```python
X_t_BN = BN.transform(X_test)
```

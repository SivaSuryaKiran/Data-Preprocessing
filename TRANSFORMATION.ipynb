# FEATURE ENGINEERING(TRANSFORMATION)

- Log Transformation
- Sqaure Root Transformation
- Cube Root Transformation
- Power Transformation
- BOX-COX Transformation
- Yeo Johnson Tranformation

```python
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
boston_dataset = load_boston()
df = pd.DataFrame(boston_dataset.data,columns=boston_dataset.feature_names)
```

```python
df.head()
```

## Plotting all columns in the dataset

```python
plt.figure(figsize=(20,25),facecolor='white')
plotnumber = 1

for column in df:
    if plotnumber<14:
        ax = plt.subplot(7,2,plotnumber)
        sns.distplot(df[column])
        plt.xlabel(column)
    plotnumber+=1
plt.show()
```

we can see the columns DIS,NOX and LSTAT are right skewed.we can perform different TRANSFORMATION techniques

## Log Transformation

### In Log Transformation , log is applied to the datapoints for selected columns

```python
df[['DIS','NOX','LSTAT']] = np.log(df[['DIS','NOX','LSTAT']])
```

```python
def plots(df, variable):
    plt.figure(figsize=(15,6))
    plt.subplot(1, 2, 1)
    sns.distplot(df[variable])
    plt.show()
```

```python
plots(df,'DIS')
```

```python
plots(df,"LSTAT")
```

## Sqaure Root transformation

### In Sqaure Root Transformation ,  Sqaure Root is applied to the datapoints 

```python
df[['DIS','NOX','LSTAT']] = np.sqrt(df[['DIS','NOX','LSTAT']])
```

```python
plots(df,'DIS')
```

```python
plots(df,'LSTAT')
```

## cube root transformation 

### In Cube Root Transformation , Cube Root is applied to the datapoints for selected columns

```python
df[['DIS','NOX','LSTAT']] = np.sqrt(df[['DIS','NOX','LSTAT']])
```

```python
plots(df,'DIS')
```

```python
plots(df,'LSTAT')
```

## Reciprocal Transformation

### In this Transformation, the values in the columns will get  reciprocal(x---> 1/x)

```python
df[['DIS','NOX','LSTAT']] = np.reciprocal(df[['DIS','NOX','LSTAT']])
```

```python
plots(df,'DIS')
```

```python
plots(df,'LSTAT')
```

## Power Transformation

### In this  Tranformation, the values of the columns are multipled by them self's with some value(x---->x^2)

```python
df[['DIS','NOX','LSTAT']] = np.power(df[['DIS','NOX','LSTAT']],0.4)
```

```python
plots(df,'DIS')
```

```python
plots(df,'LSTAT')
```

## BOX-COX Transformation

BOX-COX returns two parametes   
1) Transformed data      
2) Lambda value

```python
df['LSTAT'], parameter1 = stats.boxcox(df['LSTAT'])
```

```python
plots(df,'LSTAT')
```

```python
print("Parameter of LSTAT col:",parameter1)
```

```python
df['DIS'],parameter2 = stats.boxcox(df['DIS'])
```

```python
plots(df,'DIS')
```

```python
print("parameter of DIS col:",parameter2)
```

## Yeo Johnson Transformation

Yeo Johnson Transformation  also returns two paramaters       
1) Transformed data   
2) Lambda Value

```python
df['LSTAT'], parameter = stats.yeojohnson(df['LSTAT'])
```

```python
plots(df,'LSTAT')
```

```python
df['DIS'], parameter = stats.yeojohnson(df['DIS'])
```

```python
plots(df,'DIS')
```

# Conclusion

Out of all Transformations, BOX-COX and Yeo Johnson Transformation are best for the  given DATASET.Because we can see Tranformed data plots are looking bell shaped by using BOX-COX and Yeo Johnson transformations

# Feature Engineering(Handling Missing values)

## handling for categorical data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## reading the dataset
data = pd.read_csv(r'C:\Users\P nageswararao\OneDrive\Desktop\hr_cleaned.csv')
```

```python
## dataset information 
data.info()
```

```python
## checking for missing values
data.isnull().sum()
```

### Replacing missing value with RANDOM VALUE

```python
data['gender'] = data['gender'].replace(np.nan,'male')
```

```python
data['gender'].isnull().sum()
```

```python
data['enrolled_university'] = data['enrolled_university'].replace(np.nan,'Full time course')
```

```python
data['enrolled_university'].isnull().sum()
```

```python
data['education_level'] = data['education_level'].replace(np.nan,'Graduate')
```

```python
data['education_level'].isnull().sum()
```

```python
data['major_discipline'] = data['major_discipline'].replace(np.nan,'STEM')
```

```python
data['major_discipline'].isnull().sum()
```

```python
data['company_type']  = data['company_type'].replace(np.nan,'Small & Medium-org')
```

```python
data['company_type'].isnull().sum()
```

```python
data.isnull().sum()
```

### Replacing missing values using FORWARD FILLING

```python
data['gender'] = data['gender'].ffill()
```

```python
data['enrolled_university'] = data['enrolled_university'].ffill()
```

```python
data['education_level'] = data['education_level'].ffill()
```

```python
data['major_discipline'] = data['major_discipline'].ffill()
```

```python
data['company_type'] = data['company_type'].ffill()
```

```python
data.isnull().sum()
```

### replacing missing values with BACKWARD FILLING

```python
data['gender'] = data['gender'].bfill()
```

```python
data['enrolled_university'] = data['enrolled_university'].bfill()
```

```python
data['education_level'] = data['education_level'].bfill()
```

```python
data['major_discipline'] = data['major_discipline'].bfill()
```

```python
data['company_type'] = data['company_type'].bfill()
```

```python
data.isnull().sum()
```

### Replacing missing values with MODE

```python
data['gender'] = data['gender'].replace(np.nan,data['gender'].mode)
```

```python
data['enrolled_university'] = data['enrolled_university'].replace(np.nan,data['enrolled_university'].mode)
```

```python
data['education_level'] = data['education_level'].replace(np.nan,data['education_level'].mode)
```

```python
data['major_discipline'] = data['major_discipline'].replace(np.nan,data['major_discipline'].mode)
```

```python
data['company_type']  = data['company_type'].replace(np.nan,data['company_type'].mode)
```
